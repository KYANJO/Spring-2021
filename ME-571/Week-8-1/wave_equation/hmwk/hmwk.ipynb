# Task 2

```python
#strong scaling

#import task1 data 
task2_strong = pd.read_csv('strong.csv')
# number of processors
np=array(task2_strong['nproc'])

#serial time
T1 = array(task2_strong[' gtime'][0])

#parallel time
Tp = array(task2_strong[' gtime'])

#speed up
s = T1/Tp

#efficiency plot
eff = s/np

#ideal curve
e = s/s
```

```python
figure(1)
loglog(np,s,"-*",label='measurement')
loglog(np,np,label='ideal')
ylabel("speed up")
grid()
xlabel("number of processes")
title("speed up plot strong scaling for $N=2^{10}$ task2")
legend()
figure(2)
loglog(np,eff,"-*",label='measurement')
loglog(np,e,label='ideal')
title("Efficiency plot strong scaling for $N=2^{10}$ task2")
ylabel("efficiency")
xlabel("number of processes")
grid()
legend()
show()
```

```python
 #import task1 data
task2_weak = pd.read_csv('weak.csv') # number of processors np=array(task1_weak['nproc'])
#serial time
Tw2 = array(task2_weak[' gtime'][0]) #parallel time
Twp = array(task2_weak[' gtime']) #speed up
sw = Tw2/Twp
sw2 = np*Tw2/Twp
#efficiency plot
effw = sw2/np 
e = sw/sw
```

```python
figure(3) 
loglog(np,sw2,"-*",label='measurement') 
loglog(np,np,label='ideal')
ylabel("speed up")
xlabel("number of processes")
legend()
grid()
title("speed up plot weak scaling task2") 
figure(4) 
loglog(np,effw,"-*",label='measurement') 
loglog(np,e,label='ideal') 
title("Efficiency plot weak scaling task2") 
ylabel("efficiency")
xlabel("number of processes")
grid()
legend()
show()
```

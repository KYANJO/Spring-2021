# Task 2
## Perform strong scaling

```python
#import task2 data
task2_strong = pd.read_csv('task2_strong.csv')
T1 = array(task2_strong['gtime'][0])

# number of processors
np=array(task2_strong['nproc'])

#serial time
T1 = array(task2_strong['gtime'][0])

#parallel time
Tp = array(task2_strong['gtime'])

#speed up
s = T1/Tp

#efficiency plot
eff = s/np

#ideal curve
e = s/s
```

```python
#ploting
figure(1)
loglog(np,s,"-*",label='measurement')
loglog(np,np,label='ideal')
ylabel("speed up")
grid()
xlabel("number of processes")
title("speed up plot strong scaling for $N=840$ task2")
legend()
figure(2)
semilogx(np,eff,"-*",label='measurement')
semilogx(np,e,label='ideal')
title("Efficiency plot strong scaling for $N=840$ task2")
ylabel("efficiency")
xlabel("number of processes")
grid()
legend()
show()
```

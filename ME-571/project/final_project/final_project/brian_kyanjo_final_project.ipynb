# strong scaling

```python
2**28
```

```python
#import task1 data 
task1_strong = pd.read_csv('strong_t1.csv')
# number of processors
np=array(task1_strong['nproc'])

#serial time
T1 = array(task1_strong['time'][0])

#parallel time
Tp = array(task1_strong['time'])

#speed up
s = T1/Tp

#efficiency plot
eff = s/np

#ideal curve
e = s/s
```

```python
figure(2)
loglog(np,s,"-*",label='measurement')
loglog(np,np,label='ideal')
ylabel("speed up")
grid()
xlabel("number of processes")
title("speed up plot strong scaling for $N=2^{28}$ task1")
legend()
figure(3)
semilogx(np,eff,"-*",label='measurement')
semilogx(np,e,label='ideal')
title("Efficiency plot strong scaling for $N=2^{28}$ task1")
ylabel("efficiency")
xlabel("number of processes")
grid()
legend()
show()
```

# weak scaling

```python
#import task1 data 
task1_weak = pd.read_csv('weak_sc1.csv')
# number of processors
np=array(task1_weak['nproc'])

#serial time
Tw1 = array(task1_weak['time'][0])

#parallel time
Twp = array(task1_weak['time'])

#speed 
sw = Tw1/Twp

sw1 = np*sw

#efficiency plot
effw = sw1/np

e = sw/sw
#task1_weak
figure(5)
semilogx(np,effw,"-*",label='measurement')
semilogx(np,e,label='ideal')
title("Efficiency plot weak scaling task1")
ylabel("efficiency")
xlabel("number of processes")
grid()
legend()
show()
```

# Task 3


```python
#import task3 data
task3_1 = pd.read_csv('results3_1.csv')
task3_2 = pd.read_csv('results3_2.csv')
task3_mpi = pd.read_csv('result3_mpi.csv')
task3_se = pd.read_csv('result3_se.csv')

T1 = array(task3_1['time'])
T2 = array(task3_2['time'])
Tmpi = array(task3_mpi['time'])
Tse = array(task3_se['time'])

# problem size
N=array(task3_1['N'])

```

```python
#ploting
loglog(N,T1,'-*',label='task1_cuda')
loglog(N,T2,'-*',label='task2_cuda')
loglog(N,Tmpi,'-*',label='mpi')
loglog(N,Tse,'-*',label='serial')
title("Time plot")
ylabel("time")
xlabel("N")
grid()
legend()
show()
```

# Task 3
## Perform strong scaling

```python
#import task3 data
task3_strong = pd.read_csv('task3_strong.csv')

# number of processors
np=array(task3_strong['nproc'])

#serial time
T13 = array(task3_strong['gtime'][0])

#parallel time
Tp3 = array(task3_strong['gtime'])

#speed up
s3 = T13/Tp3

#efficiency plot
eff3 = s3/np

#ideal curve
e3 = s3/s3
```

```python
#ploting
legend()
figure(3)
semilogx(np,eff,"-o",label='2D-decomposition')
semilogx(np,eff3,"--*",label='1D-decomposition')
semilogx(np,e3,label='ideal')
title("Efficiency plot strong scaling for $N=840$")
ylabel("efficiency")
xlabel("number of processes")
grid()
legend()
show()
```

# No.2

```python
#number of processors
np = array([1,2,4,8])

#time taken to complete intialisation 
tinit = array([8.250000e+00,4.710000e+00,1.920000e+00,1.140000e+00])
#time taken to complete computation 
tcomp = array([1.900000e+00,8.800000e-01,2.800000e-01,1.200000e-01])
```

```python
#Plotting
figure(1)
loglog(np,tinit,'-*',label="initialisation_time")
loglog(np,tcomp,'-^',label="computation_time")
title("A graph of time vs number of processes")
xlabel("Number of processes")
ylabel("Time")
legend()
show()
```

Its observed that the simulation takes small time for computation than the initialiasation. Also from the results i observed that the last ranks actually use more time than the first ranks, however, still the computation time is lower than the intialsation  time for each run.

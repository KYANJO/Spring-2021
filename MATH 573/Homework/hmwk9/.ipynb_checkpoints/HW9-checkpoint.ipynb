```python
%matplotlib notebook
%pylab
import matplotlib.animation as animation
```

```python
def H(t,t0):
    return where(t-t0 >= 0, 1, 0)


def delta(x,x0,t,t0, eps):
    
    df = (1/sqrt(2*pi*eps))*exp((-(x-x0-c*(t-t0))**2)/(4*eps))
    return df

def beta(d0, eps, t0):
    
    return d0*sqrt(2*pi*eps)/(t0 + sqrt(2*pi*eps))

def u(x,x0,t,t0, eps, d0):
    
    u1 = beta(d0, eps, t0)*delta(x,x0,t,t0, eps)*(t +(t-t0)*H(t,t0))
    
    return u1
```

```python
x = arange(0, 1, 0.01)
c = 0.5
# parameters
x0 = 0.33
t0 = 0.5
# observation data
d0 = 0.68
eps = 1e-5
beta0 = beta(d0, eps, t0)

time = arange(0,0.501, 0.01)
```

```python
fig = figure(2)
u0 = u(x,x0,0,t0, eps, d0)
hdl, = plot(x,u0,'b-',markersize=5,label='Model')

tstr = 't = {:.4f}'
htitle = title(tstr.format(0),fontsize=18)

plot(x0, d0, 'r*-', label = 'Observation')

legend()

for t in time:

    ut = u(x,x0,t,t0, eps, d0)
    
    hdl.set_ydata(ut)
    htitle.set_text(tstr.format(t))
    ylim([-0.5,1.0])
    pause(0.1)

    fig.canvas.draw()
```
